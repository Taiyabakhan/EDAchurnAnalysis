```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Customer Churn.csv")
df.head()
```

```python
df.info()
```

```python
#replacing blamks with zero and conerting into float
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype(float)
```

```python
#converting 0 and 1 values into yes/no
def conv(value):
  if value==1:
    return "yes"
  else:
    return "no"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
ax=sns.countplot(x=df["Churn"],data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by churn")
plt.show()
```

```python
#group by to collect and sum the values
plt.figure(figsize=(4,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],data=df,labels=df["Churn"].unique(),autopct="%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

```python
sns.countplot(x=df["gender"],data=df)
plt.title("Count of customers by gender")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x=df["gender"],data=df,hue="Churn")
plt.title("Count of customers by gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df["SeniorCitizen"],data=df,hue="Churn")
plt.title("Count of customers by SeniorCitizen")
plt.plot()
```

```python
# Calculate counts
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate percentages
percentage_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

# Plotting
plt.figure(figsize=(5, 3))
percentage_data.plot(kind='bar', stacked=True)

# Adding percentage labels
for i in range(len(percentage_data)):
    for j in range(len(percentage_data.columns)):
        plt.text(i, percentage_data.iloc[i, :].cumsum()[j] - percentage_data.iloc[i, j] / 2,
                 f"{percentage_data.iloc[i, j]:.1f}%", ha='center', va='center')

plt.title("Percentage of Customers by Senior Citizen Status and Churn")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.xticks(ticks=range(len(percentage_data.index)), labels=percentage_data.index, rotation=0)
plt.legend(title='Churn', labels=['Not Churned', 'Churned'])
plt.tight_layout()
plt.show()
```

```python
# List of columns to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the matplotlib figure
num_cols = 3  # Number of columns in the subplot grid
num_rows = (len(columns) + num_cols - 1) // num_cols  # Calculate number of rows needed
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create count plots for each column
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i])
    axes[i].set_title(f'Count Plot of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
